```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def forward_difference(g,x,h):
    return (g(x+h) - g(x))/h
def central_difference(g,x,h):
    return (g(x+h) - g(x-h))/(2*h)
def f(x):
    return np.sin(x)    
```

```python
hs = [0.5, 0.2, 0.1, 0.05, 0.02, 0.01, 0.005, 0.002, 0.001, 0.0005, 0.0002, 0.0001]
```

```python
N = len(hs)
calculated_derivative_using_FD = np.zeros(N)
k = 0
x = 2*np.pi/5
for h in hs:
    #calculated_derivative_using_FD[k] = forward_difference(f,x,h)
    calculated_derivative_using_FD[k] = central_difference(f,x,h)
    k += 1

```

```python
ydata = np.asarray(np.log(abs(calculated_derivative_using_FD - np.cos(2*np.pi/5))))
xdata = np.asarray(np.log(hs))
plt.plot(xdata, ydata,'go')
```

```python
def eq_of_line(xs, p0, p1):
    return p0*xs+p1
```

```python
params, covariance = curve_fit(eq_of_line, xdata, ydata, p0=[1.0, 1.0])
```

```python
plt.plot(xdata, ydata,'go')
plt.plot(xdata, eq_of_line(xdata, params[0], params[1]))
```
